## 5.	How does the deal_equity distribution look? Are there any outliers

```python
deal_equities = df['deal_equity'].dropna()
plt.figure(figsize=(8, 2.5))
sns.boxplot(x=deal_equities, color='purple')

plt.title('Boxplot of Deal Equity (%)')
plt.xlabel('Deal Equity (%)')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# IQR method to find outliers
Q1 = deal_done['deal_equity'].quantile(0.25)
Q3 = deal_done['deal_equity'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = deal_done[
    (deal_done['deal_equity'] < lower_bound) |
    (deal_done['deal_equity'] > upper_bound)
]

# Show outlier brands and their deal equity
print(outliers[['brand_name', 'deal_equity']])
```

## 6.	Find the correlation between deal_valuation and ask_valuation. What insight can be drawn from this?

```python
deal_done=df[df["deal"]==1]
correlation= deal_done['deal_valuation'].corr(deal_done['ask_valuation']).round(3)
correlation
```

insights:
Correlation ≈ 1 → Strong Positive Relationship
As the ask_valuation increases, the deal_valuation also increases.
This means sharks tend to agree with the startup's valuation, even if they negotiate slightly.
Founders’ expectations are often in line with what sharks are willing to pay.

Correlation ≈ 0 → No Relationship
There's no clear pattern between ask and deal valuations.
A startup might ask for a high valuation and get a low deal — or the opposite.
This shows inconsistent investor perception or heavy negotiation, regardless of ask.

Correlation ≈ -1 → Strong Negative Relationship
As the ask_valuation increases, the deal_valuation decreases.
Indicates that sharks strongly disagree with the founders’ valuation.
They may feel the ask is unrealistic and offer lower deals, or even pull out.

## 14.	Which shark has the highest return on investment (ROI) based on deal_amount vs. deal_equity?

```python
deals_df = df[df['deal'].astype(str).str.strip() == '1'].copy()
deals_df['amount_per_shark'] = deals_df['deal_amount'] / deals_df['total_sharks_invested']

deals_df['roi'] = deals_df['deal_equity'] / deals_df['amount_per_shark']
shark_columns = [
    'ashneer_deal', 'anupam_deal', 'aman_deal',
    'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal'
]

shark_rois = {}
for shark in shark_columns:
    shark_deals = deals_df[deals_df[shark] == 1]
    avg_roi = shark_deals['roi'].mean()
    # Format shark name (capitalize first letter)
    shark_name = shark.replace('_deal', '').capitalize()
    shark_rois[shark_name] = avg_roi

sorted_shark_rois = dict(sorted(shark_rois.items(), key=lambda x: x[1], reverse=True))

print("Average ROI per Shark (Equity % per ₹ Lakh Invested):")
for shark, roi in sorted_shark_rois.items():
    print(f"{shark}: {roi:.2f}")
```

## 18.	Analyze the impact of equity dilution on deal_valuation. Are lower equity asks leading to higher deal valuations?

```python
deals_df['rounded_equity'] = deals_df['deal_equity'].round()
avg_valuation_by_equity = deals_df.groupby('rounded_equity')['deal_valuation'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_valuation_by_equity, x='rounded_equity', y='deal_valuation', marker='o', color='green')
plt.title('Average Deal Valuation vs Equity Given')
plt.xlabel('Equity Given (%)')
plt.ylabel('Average Deal Valuation (₹ lakh)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

This analysis helps us understand if equity dilution impacts valuation, and whether startups that offer less equity still command strong interest and better valuations from the sharks
